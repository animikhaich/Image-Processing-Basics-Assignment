# Take 512 x 512 image and reduce size to 50% and 80%

### Algorithm:
1. For reduction to 50%, take the image and remove every 2nd row and column from the image matrix.
2. For reduction to 80%, take the image and remove every 5th row and column from the image matrix. 

```python
# Imports
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read and display the image dimentions (Greyscale)
image = cv2.imread('rose.jpg', 0)
image.shape
```

```python
# Display the image using matplotlib, grayscale.
plt.imshow(image, cmap='gray')
```

```python
# Remove every alternative row and column from the image and display the resulting dimentions
reduced_image_50_percent = image[::2, ::2]
reduced_image_50_percent.shape
```

```python
# Display the resultant reduced to 50% size image
plt.imshow(reduced_image_50_percent, cmap='gray')
```

```python
# Remove every 5th row and column for reduction of the image to 80% of its original size and display the resulting shape
temp_img = np.delete(image, list(range(0, image.shape[0], 5)), axis=0)
reduced_image_80_percent = np.delete(temp_img, list(range(0, temp_img.shape[1], 5)), axis=1)
reduced_image_80_percent.shape
```

```python
#Display the resulting image thus formed form the above step
plt.imshow(reduced_image_80_percent, cmap='gray')
```
