# Reduce intensity resolution from 8 bits to 7, 6, 5, 4, 3, 2, 1 and display them

### Algorithm:
1. Divide each elemnt of the 'n' bit original image by 2^n
2. In order to convert to 'k' bit image, multiply the above result by 2^k-1 and round off to the nearest integer
3. After rounding off, make all the negative values 0 (Approximation)
4. Display the resulting images

```python
# Import relevant libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read the image in grayscale mode and display the maximum pixel value
image_8_bit = cv2.imread('rose.jpg', 0)
np.max(image_8_bit)
```

```python
# Display the image in it's original 8 bit form
plt.imshow(image_8_bit, cmap='gray')
```

```python
# Convert the image to 7 bit image by using the mentioned algorithm, where, n = 8, k = 7
image_7_bit = np.round((image_8_bit/256)*128-1)
image_7_bit[image_7_bit < 0] = 0

# Display all the unique pixel values, they lie between 0 and 128
np.unique(image_7_bit)
```

```python
# Display the 7 bit image
plt.imshow(image_7_bit, cmap='gray')
```
