# Perform noise removal using Average Filter, Weighted Average Filter, Gaussian Filter and Median Filter

### Algorithm:
1. Declare the filters
2. Use inbuilt function to apply convolution operation for each of the filters on the given image
3. Display the resulting images

```python
# Import the required libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import convolve2d
from scipy.ndimage import median_filter
import matplotlib.gridspec as gridspec
```

```python
# Read the image in grayscale and display it
image = cv2.imread('rose.jpg', 0)
plt.imshow(image, cmap='gray')
```

```python
# Declar the various filters to be tested
avg_filter_less_blur = np.ones((5, 5))/25

avg_filter_more_blur = np.ones((30, 30))/900

wt_avg_filter = np.array([[1, 2, 1], 
                          [2, 4, 2], 
                          [1, 2, 1]])/16

gaussian_filter = np.array([[1, 4, 7, 4, 1],
                            [4, 16, 26, 16, 4],
                            [7, 26, 41, 26, 7],
                            [4, 16, 26, 16, 4],
                            [1, 4, 7, 4, 1]])/273
```

```python
# Apply convolution operation between the image and each of the filters
less_blur_op_img = convolve2d(image, avg_filter_less_blur, mode="same", boundary="symm")
more_blur_op_img = convolve2d(image, avg_filter_more_blur, mode="same", boundary="symm")
weighted_avg_img = convolve2d(image, wt_avg_filter, mode="same", boundary="symm")
gaussian_op_img = convolve2d(image, gaussian_filter, mode="same", boundary="symm")
median_op_img_less = median_filter(image, size=(5, 5))
median_op_img_more = median_filter(image, size=(30, 30))
```

```python
# Plot the figures
plt.figure(figsize=(20,20))
gs = gridspec.GridSpec(4, 2)
gs.update(wspace=0.025, hspace=0.1)

fig = plt.subplot(gs[0])
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

fig = plt.subplot(gs[1])
plt.imshow(weighted_avg_img, cmap='gray')
plt.title("Weighted Average Filtered Image, Window Size: 3")
plt.axis('off')

fig = plt.subplot(gs[2])
plt.imshow(less_blur_op_img, cmap='gray')
plt.title("Less Blurred Average Filtered Image, Window Size: 5")
plt.axis('off')

fig = plt.subplot(gs[3])
plt.imshow(more_blur_op_img, cmap='gray')
plt.title("More Blurred Average Filtered Image, Window Size: 30")
plt.axis('off')

fig = plt.subplot(gs[4])
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

fig = plt.subplot(gs[5])
plt.imshow(gaussian_op_img, cmap='gray')
plt.title("Gaussian Filtered Image, Window Size: 5")
plt.axis('off')

fig = plt.subplot(gs[6])
plt.imshow(median_op_img_less, cmap='gray')
plt.title("Less Blurred Median filtered Image, Window Size: 5")
plt.axis('off')

fig = plt.subplot(gs[7])
plt.imshow(median_op_img_more, cmap='gray')
plt.title("More Blurred Median filtered Image, Window Size: 30")
plt.axis('off')

plt.show()
```
